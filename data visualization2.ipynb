# Bar graphs

```python
# example data

cat = ['Neha', 'kanchan','Deepika','Suhana']
values = [50,60,80,90]

# bar plot

import matplotlib.pyplot as plt

plt.bar(cat,values, color='red', width=0.8)
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('Marks')
```

# scatter plots

```python

x = ['Neha', 'kanchan','Deepika','Suhana', 'kirti', 'Shanu']
y = [50,60,80,90, 100,50]
z = [10,200,30,40,50,60]

plt.scatter(x,y,color='blue', marker='s')
plt.scatter(x,z,color='green', marker='o')

plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('Marks')
```

# Line Plot

```python

x = [1,2,3,4,5] # years
a = [10,30,60,20,50] # sales

plt.plot(x,a, label='nvidia', color='red')
```

```python
import numpy as np

x = [1,2,3,4,5] # years
a = [10,30,60,20,50] # sales
b = [5,10,50,20,15] # sales
c = [100,50,80,400,400] # sales

plt.plot(x,a, label='nvidia', color='red') 
plt.plot(x,b, label='intel', color='blue')
plt.plot(x,c, label='amd', color='green')

plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Company Sales Comparison')
plt.legend()
plt.show()
```

# Histogram

```python
import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
iris.head()
iris.info()
iris.describe()
```

```python
sns.histplot(data=iris, x='petal_length', bins=100, kde=True, color='blue')
```

```python
sns.kdeplot(data=iris, x='petal_length', hue='species', color='blue',  fill=True, alpha=0.5)
```

```python
column = ['species']

# count of each species
iris['species'].value_counts() 
```

```python

# Skewedness check
sns.countplot(x='species', data= iris)
```

# Pairplot

```python
sns.pairplot(iris, markers=["o", "s","p"], hue='species', palette='husl')
```
