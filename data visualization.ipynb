```python
import matplotlib.pyplot as plt
import numpy as np

#sine wave parameters

x = np.linspace(0, 10, 100)

y = np.sin(x)
y2 = np.sin(x + np.pi / 4)
y3 = np.cos(x)
#y4 = np.tan(x)


plt.plot(x, y, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Sine Wave + π/4', color='orange')
plt.plot(x, y3, label='Cosine Wave', color='green')
#plt.plot(x, y4, label='Tangent Wave', color='red')

plt.title('Sine, Cosine, and Tangent Waves')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sine_wave(frequency, duration, sample_rate):
    """
    Plots a sine wave with the given frequency, duration, and sample rate.

    Parameters:
    frequency (float): Frequency of the sine wave in Hz.
    duration (float): Duration of the sine wave in seconds.
    sample_rate (int): Number of samples per second.

    Returns:
    None
    """
    # Generate time values
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    
    # Generate sine wave values
    y = np.sin(2 * np.pi * frequency * t)
    
    # Plot the sine wave
    plt.figure(figsize=(10, 4))
    plt.plot(t, y)
    plt.title(f'Sine Wave: {frequency} Hz')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.show()
```

```python
# plot a sine wave with frequency 440 Hz, duration 1 second, and sample rate 44100 Hz
plot_sine_wave(5, 1, 1000)
```
